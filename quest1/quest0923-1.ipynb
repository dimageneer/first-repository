```python
import re

# 원본 텍스트 파일을 읽습니다.
with open('Avengers.txt', 'r', encoding='utf-8') as file:
    original_text = file.read()

# 공백과 구두점을 구분자로 사용하여 텍스트를 단어로 분할합니다.
words = re.findall(r'\b\w+\b', original_text.lower())

# 단어의 개수를 셉니다.
word_count = len(words)

print("단어 개수:", word_count)
```

```python
# 비문자 제거된 텍스트를 소문자로 변환합니다.
text_lowercase = text_only_letters.lower()

print("소문자로 변환된 내용:")
print(text_lowercase)
```

```python
from collections import Counter

# 단어 빈도수를 계산합니다.
word_counts = Counter(words)

# 빈도수 순으로 내림차순 정렬하여 출력합니다.
for word, count in word_counts.most_common():
    print(f"{word}: {count}")
```

```python
# 빈도수 순으로 정렬된 단어 리스트를 가져옵니다.
sorted_words = [word for word, count in word_counts.most_common()]

# 단어와 인덱스를 매핑하는 딕셔너리를 생성합니다.
word_to_index = {word: index for index, word in enumerate(sorted_words)}

print("단어-인덱스 매핑 (빈도수 순):")
print(word_to_index)
```

```python
# 인덱스가 부여된 단어들 (word_to_index 딕셔너리의 키)을 출력합니다.
print("인덱스된 단어 목록:")
print(list(word_to_index.keys()))
```
